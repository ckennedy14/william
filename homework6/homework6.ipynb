```python
lst = [20, 27, 9, 12, 31, 17, 4, 50, 39, 35]

arr = np.array(lst)

print(arr * 4)
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))

min_index = np.argmin(arr)
max_index = np.argmax(arr)

print(min_index)
print(max_index)

print(arr[min_index])
print(arr[max_index])

print(np.min(arr))
print(np.max(arr))

print(arr[arr > 20])
```

```python
matrix = np.array([
    [5, 12, 17, 9],
    [8, 21, 14, 3],
    [6, 13, 19, 11]
])

print(matrix.shape)
print("Rows:", matrix.shape[0])
print("Columns:", matrix.shape[1])

print(matrix[:, 2])
print(matrix[1, :])

print(matrix * 2)

print(np.mean(matrix, axis=0))
print(np.mean(matrix))

print(np.max(matrix, axis=1))
print(np.min(matrix, axis=0))

flat = matrix.flatten()
print(flat)
print(flat.shape)

min_flat_index = np.argmin(matrix)
max_flat_index = np.argmax(matrix)

print(min_flat_index)
print(max_flat_index)

print(np.unravel_index(min_flat_index, matrix.shape))
print(np.unravel_index(max_flat_index, matrix.shape))

print(matrix[matrix > 15])

print(np.sort(matrix, axis=1))
```

```python
arr = np.array([
    [2, 3, 5],
    [14, 6, 8],
    [11, 13, 17],
    [7, 10, 13]
])

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def prime_rows(array):
    keep_rows = []
    for row in array:
        has_prime = False
        for value in row:
            if is_prime(value):
                has_prime = True
        if has_prime:
            keep_rows.append(row)
    return np.array(keep_rows)

print(prime_rows(arr))
```

```python
def checkerboard():
    board = np.zeros((8, 8), dtype=int)
    board[0::2, 0::2] = 1
    board[1::2, 1::2] = 1
    return board

print(checkerboard())
```

```python
def reverse_checkerboard():
    board = np.zeros((8, 8), dtype=int)
    board[0::2, 1::2] = 1
    board[1::2, 0::2] = 1
    return board

print(reverse_checkerboard())
```

```python
def expansion(words, spaces):
    gap = " " * spaces
    expanded = []
    for word in words:
        expanded.append(gap.join(list(word)))
    return np.array(expanded)

universe = np.array(["galaxy", "clusters"])
print(expansion(universe, 1))
print(expansion(universe, 2))
```

```python
def secondDimmest(stars):
    result = []
    for row in stars:
        sorted_row = np.sort(row)
        result.append(sorted_row[-2])
    return np.array(result)

np.random.seed(123)
stars = np.random.randint(500, 2000, (5, 5))

print(stars)
print(secondDimmest(stars))
```
