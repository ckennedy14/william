```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
```

```python
x = np.linspace(0, 2 * np.pi, 1000)

amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(25, 5))

linestyles = ["-", "--", "-.", ":", "-"]

for A, style in zip(amplitude, linestyles):
    y = make_sine_wave(x, A, 1)
    axes[0].plot(x, y, linestyle=style, label=f"A={A}, w=1")

axes[0].set_title("Varying Amplitude", fontsize=16)
axes[0].set_xlabel("x", fontsize=14)
axes[0].set_ylabel("y", fontsize=14)
axes[0].legend()

for w, style in zip(frequency, linestyles):
    y = make_sine_wave(x, 1, w)
    axes[1].plot(x, y, linestyle=style, label=f"A=1, w={w}")

axes[1].set_title("Varying Frequency", fontsize=16)
axes[1].set_xlabel("x", fontsize=14)
axes[1].set_ylabel("y", fontsize=14)
axes[1].legend()

for A, w, style in zip(amplitude, frequency, linestyles):
    y = make_sine_wave(x, A, w)
    axes[2].plot(x, y, linestyle=style, label=f"A={A}, w={w}")

axes[2].set_title("Varying Amplitude and Frequency", fontsize=16)
axes[2].set_xlabel("x", fontsize=14)
axes[2].set_ylabel("y", fontsize=14)
axes[2].legend()

plt.tight_layout()
plt.show()
```
