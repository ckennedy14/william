```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows=1)
print(data.shape)
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]

print("Minimum magnitude:", np.min(magnitude))
print("Maximum magnitude:", np.max(magnitude))

print("Minimum temperature:", np.min(temperature))
print("Maximum temperature:", np.max(temperature))
```

```python
brightest_index = np.argmin(magnitude)
dimmest_index = np.argmax(magnitude)

hottest_index = np.argmax(temperature)
coldest_index = np.argmin(temperature)

print("Brightest star index:", brightest_index)
print("Dimmest star index:", dimmest_index)
print("Hottest star index:", hottest_index)
print("Coldest star index:", coldest_index)
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(log_temperature, log_magnitude, label="Stars", alpha=0.7)

plt.scatter(log_temperature[brightest_index], log_magnitude[brightest_index],
            color="black", marker="*", s=300, label="Brightest Star")

plt.scatter(log_temperature[dimmest_index], log_magnitude[dimmest_index],
            color="purple", marker="*", s=300, label="Dimmest Star")

plt.scatter(log_temperature[hottest_index], log_magnitude[hottest_index],
            color="green", marker="*", s=300, label="Hottest Star")

plt.scatter(log_temperature[coldest_index], log_magnitude[coldest_index],
            color="orange", marker="*", s=300, label="Coldest Star")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.title("Hertzsprung-Russell Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")
plt.legend()

plt.show()
```
